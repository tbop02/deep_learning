### Cosine similarity calculation

```python
from sklearn.preprocessing import normalize
sci_row=job_lan.loc['Data scientist or machine learning specialist'].values
sim_with_DS=np.zeros((job_lan.shape[0],1))

x=sci_row
for i in range(job_lan.shape[0]):   
    y=job_lan.iloc[i].values
    #calculation cosin similarity of each job with DS
    sim_with_DS[i,:]= (np.dot(x,y) / (np.linalg.norm(x)*np.linalg.norm(y)))
## creation Pd serios of cosin similarity between DS and others jobs.
sim_with_DS_pd=pd.Series(sim_with_DS.flatten(), index=job_lan.index.values)
sim_with_DS_pd_sorted=np.abs(sim_with_DS_pd.sort_values()-1)[::-1]
```

```python
diff_sim=[0]
diff=0
for i in range(len(sim_with_DS_pd_sorted)-1):
    diff=sim_with_DS_pd_sorted[i+1] - sim_with_DS_pd_sorted[i]
    diff_sim.append(diff)
diff_sim_norm=diff_sim/sum(diff_sim)
```

### Ward clustering based on pairwise cosine similarity

## ward method but using column nomalized data (In my opinion, this is more reasonable)

```python
from scipy.cluster.hierarchy import ward, dendrogram

norm_row_job_lan = job_lan.div(job_lan.sum(axis=1), axis=0)

linkage_matrix = ward(norm_row_job_lan)
# linkage_matrix = ward(dist)

fig, ax = plt.subplots(figsize=(15, 20)) # set size
ax = dendrogram(linkage_matrix, orientation="right", labels=job_lan.index.values);

plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off',
    labelsize=100)

plt.tick_params(axis='y', which='major', labelsize=20)
plt.tight_layout() #show plot with tight layout


#uncomment below to save figure
plt.savefig('ward_clusters.png', dpi=200) #save figure as ward_clusters
```

### choose best cluster number based on silhouette score

```python
from sklearn import metrics
for i in range(2,15):
    assignments = scipy.cluster.hierarchy.fcluster(linkage_matrix, i, criterion="maxclust")
    print(metrics.silhouette_score(norm_row_job_lan, assignments, metric='euclidean'))
```

# How data scientists have been educated

### unique education types

```python
edu_list=[]
for num,i in enumerate(data.EducationTypes.dropna()):             
    for j in i.split(';'):            
            edu_list.append(j)
            edu_list=list(set(edu_list))
print(edu_list)
```

### Education Type count only regarding Data scientist

```python
DS_edu = data[data.DevType == 'Data scientist or machine learning specialist'].EducationTypes.dropna()
DS_edu_frame=pd.DataFrame(0,index=[0],columns=edu_list)
for i in DS_edu:
    for j in i.split(';'):
        for edu_ele in edu_list:
            if edu_ele == j:
                DS_edu_frame[edu_ele][0]+=1
```

```python
mpl.rcParams['font.size'] = 20

ax=pd.DataFrame(DS_edu_frame.iloc[0,:].sort_values(ascending=False)).T.plot.bar(title="How Data scientists learn to be a data scientist (only considered people who chose only data scientist)", figsize=(20,20))
ax.legend(loc='upper right')
ax.set_ylim([0,800])
```

### Education Type calculation of whole job 

```python
## data frame creationd
jobs_edu_frame=pd.DataFrame(0,index=jobs,columns=edu_list)
data_cols=data[['DevType','EducationTypes']].dropna()
for job_ele in jobs:
    for edu_ele in edu_list:
        print(job_ele,edu_ele,'complete')

        for i in range(data_cols.shape[0]):        
            if job_ele == data_cols.iloc[i].DevType:
                if edu_ele in data_cols.iloc[i].EducationTypes:
                    jobs_edu_frame[edu_ele][job_ele]+=1

## result file save
jobs_edu_frame.to_csv('jobs_edu_frame.csv')
```
